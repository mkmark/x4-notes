```python
with open("7.xml", "r") as f:
  lines = f.readlines()

import re

macros_cnt= {}

for l in lines:
  res = re.search(r"macro=\"(.*)_macro", l)
  if res:
    macro = res.group(1)
    if macro in macros_cnt:
      macros_cnt[macro] += 1
    else:
      macros_cnt[macro] = 1

macros_cnt
```
